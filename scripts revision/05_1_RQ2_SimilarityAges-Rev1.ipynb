#### Run analyses for each age group

```python
current_array_filename = ['../data/ts_array_shen268_regional_average_ages13-99.npy',
                         '../data/ts_array_shen268_regional_average_ages8-13.npy',
                         '../data/ts_array_shen268_regional_average_ages5-8.npy',
                         '../data/ts_array_shen268_regional_average_ages3-5.npy']

n_age_groups = len(current_array_filename)
n_regions    = 268
res_by_group = []
all_res      = np.zeros((n_regions, n_age_groups))

for current_group in range(n_age_groups):
    withingroupres = []
    
    ts_array = np.load(current_array_filename[current_group])    
    n_subjs = ts_array.shape[0]
    print(n_subjs)

    ts_data = np.swapaxes(ts_array, 0, 2)
    ts_data = np.swapaxes(ts_data, 0, 1)
    
    for curr_region in range(n_regions):
                    curr_data = np.squeeze(ts_data[:, curr_region, : ])

                    curr_res = partly_isc(curr_data, 
                                          pairwise = False, # True,  #make true for stats --- though not perfect, at least better
                                          summary_statistic = np.mean,
                                          verbose = False)

                    all_res[curr_region,  current_group] = curr_res;
                    
                    curr_res = partly_isc( curr_data, pairwise = True, summary_statistic = None, verbose = False)
                    withingroupres.append(curr_res[:,0])
                             
    #print('\n--------------\nnextgroup\n-------------\n')
    withingroupres = np.asarray(withingroupres)
    res_by_group.append(withingroupres)
```

### Mentalizing only

```python
current_array_filename = ['../data/ts_array_mentalizingfine_regional_average_ages13-99.npy',
                          '../data/ts_array_mentalizingfine_regional_average_ages8-13.npy',
                          '../data/ts_array_mentalizingfine_regional_average_ages5-8.npy',
                          '../data/ts_array_mentalizingfine_regional_average_ages3-5.npy']

from statsmodels.stats.multitest import multipletests

n_age_groups = len(current_array_filename)


###START FOR
for current_group in range(n_age_groups):
    ts_array = np.load(current_array_filename[current_group])
    

    ts_data = np.swapaxes(ts_array, 0, 2)
    ts_data = np.swapaxes(ts_data, 0, 1)
    ts_data.shape
    ISC, p, distribution = partly_phaseshift_isc(ts_data, 
                                               pairwise=False, #can use pairwise or ind2group
                                               summary_statistic=np.mean,
                                               n_shifts = 1000, #can reduce to speed up, better increase to 1000
                                               return_distribution=True)

    ISC[np.isnan(ISC)] = 0

    alpha = .05
    q = multipletests(p[0, :], method='fdr_bh')[1]
    corrected = q[np.newaxis, :] < alpha
    ISC[corrected == False] = 0

    print(ISC)
```
